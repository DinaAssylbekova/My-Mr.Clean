```python
import requests
import re              # module for regular expressions
import seaborn as sns
```

```python
# Getting the data
url = "https://en.wikipedia.org/w/api.php"  
res = requests.get(url, params = {"action":"query", "prop":"extracts", "titles":"Ozone_layer", "format":"json"})
data = res.json()
print(data)
```

## Cleaning

```python
raw_text = data['query']['pages']['22834']['extract']
raw_text
```

```python
def cleanhtml(raw_html):             # to remove html tags
    html = re.compile('<.*?>')
    clean = re.sub(html, '', raw_html) #re.sub()takes as input:1.sub-string to replace,2.sub-string to replace with,3.actual string
    return clean
```

```python
no_html_text = cleanhtml(raw_text)
no_html_text
```

```python
def merge_contents(data):
    splitted_text = re.split("\. |, |\.|\n| |-|\'", data)
    return splitted_text

merge_content = merge_contents(no_html_text)
merge_content
```

```python
# Tokenization
def tokenize(content):
    special_char_removed = [word for word in content if word.isalnum()] # isalnum() checks whether all the characters in a given 
                                                                  # string is alphanumeric (either a letter or a number) or not.
    num_removed = [word for word in special_char_removed if not word.isdigit()] # isdigit() returns True if all characters in a 
                                                                                # string are digits. If not, it returns False.
                                                                                
    return num_removed

collection = tokenize(merge_content)
collection
```

```python
def lower_collection(collection):
    return [word.lower() for word in collection]

plain_text = lower_collection(collection)
plain_text
```

## Term Frequency

```python
def count_frequency(collection):
    frequency = {}
    for word in collection:
        if word in frequency:
            frequency[word] +=1
        else:
            frequency[word] = 1
    
    return frequency

frequencies = count_frequency(plain_text)
frequencies
```

```python
frequencies = {k:v for k,v in sorted(frequencies.items(), key=lambda item: item[1], reverse=True)}
frequencies
```

```python
list(frequencies.items())[:10]
```

```python
def print_most_frequent(frequencies, n):
    frequencies = {k:v for k,v in sorted(frequencies.items(), key=lambda item: item[1], reverse=True)}
    return list(frequencies.items())[:n]

print_most_frequent(frequencies, 10)
```

## Visualizing

```python
xx = [v for k,v in frequencies.items()]
yy = [k for k,v in frequencies.items()]
sns.barplot(x=xx[0:20], y=yy[0:20])       #to plot horizontally
```

## Filtering

```python
import nltk
from nltk.corpus import stopwords
```

```python
stop_words = set(stopwords.words('english')) 
```

```python
def remove_stop_words(words, stop_words):
    meaningful_text = [word for word in words if word not in stop_words]
    return meaningful_text

filtered_collection = remove_stop_words(plain_text, stop_words)
filtered_collection
```

```python
freq_of_meaningful_words = count_frequency(filtered_collection)
```

```python
freq_of_meaningful_words = {k:v for k,v in sorted(freq_of_meaningful_words.items(), key=lambda item: item[1], reverse=True)}
freq_of_meaningful_words
```

```python
# Visualize in a historgram the 25 most frequent words in the filtered collection
xxx = [v for k,v in freq_of_meaningful_words.items()]
yyy = [k for k,v in freq_of_meaningful_words.items()]
sns.barplot(x=xxx[0:25], y=yyy[0:25])
```
